## Histogram stretch

```python
# Load the image
img = cv.imread(r'C:\Users\Ma548\Desktop\PythonHub\Uni Projects\DIP Project 1\Photos\low_contrast.png', 0)
img
```

```python
# Calculate histogram using cv2.calcHist()
hist = cv.calcHist([img], [0], None, [256], [0, 256])


# show the image
cv.imshow('before', img)

cv.waitKey(0)
cv.destroyAllWindows()
# Display the histogram
plt.plot(hist)
```

```python
# Create zeros array to store the stretched image
minmax_img = np.zeros((img.shape[0], img.shape[1]), dtype='uint8')

# Loop over the image and apply Min-Max formulae
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        minmax_img[i, j] = 250*(img[i, j]-np.min(img)) / \
            (np.max(img)-np.min(img))

# Displat the stretched image
# show the image
cv.imshow('before', img)
cv.imshow('after', minmax_img)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Calculate histogram using cv2.calcHist()
hist = cv.calcHist([minmax_img], [0], None, [256], [0, 256])

# Display the histogram
plt.plot(hist)
```
